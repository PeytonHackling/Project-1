```python
def combine_header_files(file_paths1, file_paths2) -> list[str]:
    """Combine two lists of file paths, ensuring no duplicates."""
    combined_files = set(file_paths1) | set(file_paths2)
    return list(combined_files)
```

```python
def load_header_data_from_multiple_files(file_paths: list[str], chunk_size: int = 100000) -> pd.DataFrame:
    all_chunks = []

    for file_path in file_paths:
        print(f"Reading: {file_path}")

        for chunk in pd.read_csv(file_path, chunksize=chunk_size, dtype=str, low_memory=False):
            all_chunks.append(chunk)

    if not all_chunks:
        raise ValueError("No header data was loaded.")
    
    combined_df = pd.concat(all_chunks, ignore_index=True)
    return combined_df
```

```python
def standardize_strings(df):
    df = df.copy()

    for col in df.columns:
        df[col] = df[col].astype("string").str.strip()
        df[col] = df[col].replace({
            "": pd.NA,
            "null": pd.NA,
            "NULL": pd.NA,
            "nan" : pd.NA,
            "N/A": pd.NA,
            "n/a": pd.NA,
        })

    return df

headerdf = standardize_strings(headerdf)
print("String columns standardized.")
```

```python

def convert_header_types_in_place(df, chunk_size=250_000):
    n = len(df)

    numeric_cols = [
        "identifier",
        "manifest_quantity",
        "weight",
        "measurement",
        "conveyance_id"
    ]

    date_cols = [
        "estimated_arrival_date",
        "actual_arrival_date"
    ]

    # ---------- numeric columns ----------
    for col in numeric_cols:
        if col not in df.columns:
            continue

        # Pre-create an empty numeric array
        out = np.empty(n, dtype="float64")

        for start in range(0, n, chunk_size):
            end = min(start + chunk_size, n)

            out[start:end] = pd.to_numeric(
                df[col].iloc[start:end],
                errors="coerce",
                downcast="integer"
            )

        df[col] = out

    # ---------- date columns ----------
    for col in date_cols:
        if col not in df.columns:
            continue

        out = np.empty(n, dtype="datetime64[ns]")

        for start in range(0, n, chunk_size):
            end = min(start + chunk_size, n)

            out[start:end] = pd.to_datetime(
                df[col].iloc[start:end],
                errors="coerce",
                cache=True
            )

        df[col] = out

    return df




convert_header_types_in_place(headerdf)
print("Data types converted safely without copying.")

```

```python
def normalize_header_text(df, chunk_size=1_000_000):
    upper_cols = [
        "carrier_code",
        "vessel_country_code",
        "manifest_unit",
        "weight_unit",
        "measurement_unit",
        "record_status_indicator",
        "in_bond_entry_type"
    ]

    title_cols = [
        "port_of_unlading",
        "foreign_port_of_lading",
        "place_of_receipt",
        "port_of_destination",
        "foreign_port_of_destination"
    ]

    chunks = []
    n = len(df)

    for start in range(0, n, chunk_size):
            end = min(start + chunk_size, n)
            chunk = df.iloc[start:end].copy()

            for col in upper_cols:
                if col in chunk.columns:
                    chunk[col] = chunk[col].str.upper()

            for col in title_cols:
                if col in chunk.columns:
                    chunk[col] = chunk[col].str.title()

            chunks.append(chunk)
    return pd.concat(chunks, ignore_index=True) 

headerdf = normalize_header_text(headerdf)
print("Text columns normalized.")
```

```python
def convert_weight_to_kg(df):
    df = df.copy()

    if "weight" in df.columns and "weight_unit" in df.columns:
        return df 
    
    df["weight"] = pd.to_numeric(df["weight"], errors="coerce")
    units =df["weight_unit"].astype("string").str.strip().str.lower()

    conversion_map = {
        "kilogram": 1.0,
        "kilograms": 1.0,
        "kg": 1.0,
        "kgs": 1.0,
        "pound": 0.453592,
        "pounds": 0.453592,
        "lb": 0.453592,
        "lbs": 0.453592,
        "ton": 907.185,
        "tons": 907.185,
        "tonne": 1000.0,
        "tonnes": 1000.0,
        "metric ton": 1000.0,
        "metric tons": 1000.0,
        "short ton": 907.185,
        "short tons": 907.185,
    }

    factors = units.map(conversion_map)

    df["weight"] = np.where(
        df["weight"].notna() & factors.notna(),
        df["weight"] * factors,
        df["weight"]
    )

    df["weight"] = pd.to_numeric(df["weight"], errors="coerce").round(2)
    df["weight_unit"] = np.where(df["weight"].notna(), "KILOGRAMS", df["weight_unit"])
    return df

headerdf = convert_weight_to_kg(headerdf)
print("Weight converted to kilograms.")
```

```python


def combine_secondary_notify_parties_in_place(df, sep=" | "):
    cols = [
        c for c in (
            f"secondary_notify_party_{i}" for i in range(1, 11)
        )
        if c in df.columns
    ]

    if not cols:
        df["second_party"] = pd.NA
        return df

    # Pull columns as NumPy arrays once (CRITICAL for speed)
    arrays = [df[c].to_numpy(copy=False) for c in cols]

    n = len(df)
    out = np.empty(n, dtype=object)
    out[:] = pd.NA

    sep_local = sep
    for i in range(n):
        parts = []

        for arr in arrays:
            val = arr[i]
            if val is not None and val is not pd.NA:
                s = str(val).strip()
                if s:
                    parts.append(s)

        if parts:
            out[i] = sep_local.join(parts)

    df["second_party"] = pd.Series(out, index=df.index, dtype="string")
    return df



combine_secondary_notify_parties_in_place(headerdf)
print("Secondary notify parties combined safely.")
    
    
```

```python
def remove_duplicates(df, primary_key):
    if primary_key in df.columns:
        df = df.drop_duplicates(subset=primary_key, keep="first")
    return df

headerdf = remove_duplicates(headerdf, HEADER_PRIMARTY_KEY)
print("Duplicates removed based on primary key.")
```

```python

def build_silver_header(df):
    silver_cols = [
        "identifier",
        "carrier_code",
        "vessel_country_code",
        "vessel_name",
        "port_of_unlading",
        "foreeign_port_of_lading",
        "foreign_port_of_lading_qualifier",
        "estimated_arrival_date",
        "manifest_quantity",
        "manifest_unit",
        "weight",
        "weight_unit",
        "measurement",
        "measurement_unit",
        "record_status_indicator",
        "place_of_receipt",
        "port_of_destination",
        "foreign_port_of_destination",
        "foreign_port_of_destination_qualifier",
        "conveyance_id_qualifier",
        "conveyance_id",
        "in_bond_entry_type",
        "mode_of_transportation",
        "second_party",
        "actual_arrival_date"
    ]

    existing_cols = [col for col in silver_cols if col in df.columns]

    # NO copy here — cheap column projection
    return df.loc[:, existing_cols]

silver_headerdf = build_silver_header(headerdf)
print("Silver header dataframe built.")
```

```python
def format_header_dates_for_output(df):

    date_cols = ["estimated_arrival_date", "actual_arrival_date"]

    for col in date_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce").dt.strftime("%Y-%m-%d")

    return df


format_header_dates_for_output(silver_headerdf)
print("Date columns formatted for output.")
```

```python
def save_silver_header(df, output_path):
    output_file = Path(output_path)
    output_file.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(output_file, index=False)

save_silver_header(silver_headerdf, SILVER_OUTPUT_PATH)

print(f"Silver header dataframe saved to {SILVER_OUTPUT_PATH}.")
```
